```python
import os
import pandas as pd
from atomsci.ddm.utils import struct_utils as su
```

```python
meas="MMP"
dir="up"
og_df = pd.read_csv(f'../AMPL_models/missing_rows_mmp.csv')

og_df
```

```python
new = pd.read_csv('../AMPL_models/datasets/scaled_descriptors/MMP_hits_up_class_cur_with_moe_descriptors (1).csv')
new
```

```python
# Assuming dataframes are df1 and df2, column is 'base_rdkit_smiles'
set1 = set(og_df['base_rdkit_smiles'].dropna())
set2 = set(new['base_rdkit_smiles'].dropna())

if set1 == set2:
    print("✅ SMILES columns are IDENTICAL")
else:
    print("❌ SMILES columns DIFFER")
    print(f"Only in df1: {set1 - set2}")
    print(f"Only in df2: {set2 - set1}")
```

```python
# remove duplicate smiles

new_clean_smiles = new[~new.duplicated(subset=["base_rdkit_smiles"], keep=False)]
new_clean_smiles.shape


# remove duplicate compound ids
new_clean = new_clean_smiles[~new_clean_smiles.duplicated(subset=["compound_id"], keep=False)]
new_clean.shape
```

```python
# merge old and new dataframes
new_MMP = pd.concat([og_df, new_clean], axis=0, ignore_index=True)
new_MMP
```

```python
# double check for duplicates
clean_MMP = new_MMP[~new_MMP.duplicated(subset=["base_rdkit_smiles"], keep=False)]
clean_MMP = clean_MMP[~clean_MMP.duplicated(subset=["compound_id"], keep=False)]

clean_MMP.shape
```

```python
# new_clean["base_rdkit_smiles"].is_unique
new_clean["compound_id"].is_unique
```

```python
new_clean.to_csv(f'/Users/echun/repos/DILI-rotation/AMPL_models/datasets/MMPclean_cur2.csv', index=False)
```

```python
# check for broken smiles, replace them

new_smi = []

for smi in clean_MMP.base_rdkit_smiles:
    try:
        new_smiles = su.base_smiles_from_smiles(smi)
    except:
        print(f"base smiles failed for {smi}")
        new_smiles = ''
    new_smi.append(new_smiles)
clean_MMP['new_smiles'] = new_smi

clean_MMP = clean_MMP[clean_MMP.new_smiles != '']
```

```python
# save to csv
new_MMP.to_csv('/Users/echun/repos/DILI-rotation/datasets_ST/MMP/MMP_hits_up_class_cur.csv', index=False)
```

```python
# df = pd.read_csv(f'/Users/echun/repos/DILI-rotation/datasets_ST/MMP/MMP_hits_up_class_cur.csv')

clean_MMP['base_rdkit_smiles'] = clean_MMP['new_smiles'].replace('', pd.NA).fillna(clean_MMP['base_rdkit_smiles'])
clean_MMP = clean_MMP.drop(columns=['new_smiles'])

# df.to_csv('/Users/echun/repos/DILI-rotation/datasets_ST/MMP/MMP_hits_up_class_cur.csv', index=False)
```
